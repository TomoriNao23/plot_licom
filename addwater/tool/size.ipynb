```python
from PIL import Image
import matplotlib.pyplot as plt
import os
```

```python
class act:
    
    def __init__(self,path,rect):
        self.p = path
        self.r = rect
        self.tem = "./tem.png"
        self.figs = (15, 15)

    def imageshow(self):
        img_out = Image.open(self.tem)
        plt.figure(figsize = self.figs)
        plt.axis('off')
        plt.imshow(img_out)
        plt.show()

    def size_change(self):
        img = Image.open(self.p)
        print(img.size)
        crop_image = img.crop(self.r)
        crop_image.save(self.tem)
        self.imageshow()

    def save(self):
        os.system("cp ./tem.png "+self.p)
        os.system("rm ./tem.png")
        
```

```python
act0 =act("addwater/photo/heat.png",(0, 355, 1024, 665))
act0.size_change()
act0.save()
```

```python
act1 = act("addwater/photo/heat_lat.png",(0, 340, 1024, 680))
act1.size_change()
act1.save()
```

```python
act2 = act("addwater/photo/delta_psi.png",(0, 135, 1024, 805))
act2.figs = (10, 10)
act2.size_change()
act2.save()
```
